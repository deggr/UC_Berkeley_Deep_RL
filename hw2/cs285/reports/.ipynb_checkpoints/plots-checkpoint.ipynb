```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 3

```python
sb1 = pd.read_csv('problem_3/run_pg_sb_no_rtg_dsa_CartPole-v0_25-12-2019_14-38-20-tag-Eval_AverageReturn.csv')
sb2 = pd.read_csv('problem_3/run_pg_sb_rtg_dsa_CartPole-v0_25-12-2019_14-40-14-tag-Eval_AverageReturn.csv')
sb3 = pd.read_csv('problem_3/run_pg_sb_rtg_na_CartPole-v0_25-12-2019_14-41-31-tag-Eval_AverageReturn.csv')

lb1 = pd.read_csv('problem_3/run_pg_lb_no_rtg_dsa_CartPole-v0_25-12-2019_14-42-46-tag-Eval_AverageReturn.csv')
lb2 = pd.read_csv('problem_3/run_pg_lb_rtg_dsa_CartPole-v0_25-12-2019_14-45-53-tag-Eval_AverageReturn.csv')
lb3 = pd.read_csv('problem_3/run_pg_lb_rtg_na_CartPole-v0_25-12-2019_14-49-09-tag-Eval_AverageReturn.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
ax1.plot(sb1['Value'])
ax1.plot(sb2['Value'])
ax1.plot(sb3['Value'])
ax1.legend(['sb no rtg dsa', 'sb rtg dsa', 'sb rtg na'])
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Avg. eval. return')

ax2.plot(lb1['Value'])
ax2.plot(lb2['Value'])
ax2.plot(lb3['Value'])
ax2.legend(['lb no rtg dsa', 'lb rtg dsa', 'lb rtg na'])
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Avg. eval. return')

plt.show()
```
